```python
# Drop duplicate rows
df = df.drop_duplicates()
print(f"After removing duplicates: {df.shape[0]} rows")
```

```python
# Remove outliers using IQR method for each numeric column
def remove_outliers_iqr(dataframe, columns):
    for col in columns:
        Q1 = dataframe[col].quantile(0.25)
        Q3 = dataframe[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        before = dataframe.shape[0]
        dataframe = dataframe[(dataframe[col] >= lower) & (dataframe[col] <= upper)]
        after = dataframe.shape[0]
        print(f"{col}: removed {before - after} outliers")
    return dataframe

# Remove outliers from all numeric features except the target ('strength')
features = df.columns.drop('strength')
df_clean = remove_outliers_iqr(df, features)
```

```python
# Apply StandardScaler to all features (except 'strength', which is typically the target)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_clean[features])

# Replace original feature columns with scaled values
df_clean[features] = X_scaled

print("Data cleaned and standardized:")
print(df_clean.head())
```
